```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
!pwd
```

```python
img = plt.imread('Nightwing_1.jpg')
```

```python
def plot_image(image: np.ndarray, title: str):
  plt.imshow(image)
  plt.title(title)
  plt.axis('off')
  plt.show()
```

```python
plot_image(img, 'Image')
```

```python
print(type(img))
print(img.shape)
print(img.ndim)
print(img.size)
print(img.dtype)
```

```python
grey_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plot_image(grey_image, 'Gray Image')
```
